```python
# Como √© a distribui√ß√£o racial das v√≠timas?

from sqlalchemy import create_engine
import pandas as pd

engine = create_engine('postgresql+psycopg2://postgres:123@localhost/postgres')
conn = engine.connect()

query = """
SELECT race AS ra√ßa, COUNT(*) AS total
FROM (
    SELECT CASE WHEN (race IS null OR race LIKE 'O') THEN 'desconhecida' ELSE race END AS race
    FROM projeto.fatal_shootings
    ) AS subquery
GROUP BY race
ORDER BY total DESC;
"""
df = pd.read_sql_query(query, con = conn)
df
```

## Consulta 2: Armas Utilizadas

Vamos explorar as armas utilizadas nas ocorr√™ncias.

### Perguntas:
- Quais s√£o as armas mais comuns envolvidas nas ocorr√™ncias?
- Qual √© a propor√ß√£o de suspeitos armados e desarmados nas ocorr√™ncias?

```python
# Quais s√£o as armas mais comuns envolvidas nas ocorr√™ncias?

from sqlalchemy import create_engine
import pandas as pd

engine = create_engine('postgresql+psycopg2://postgres:123@localhost/postgres')
conn = engine.connect()

query = """
SELECT armed AS arma, COUNT(armed) AS total
FROM projeto.fatal_shootings
GROUP BY armed
ORDER BY total DESC
LIMIT 10;
"""
df = pd.read_sql_query(query, con = conn)
df
```

As armas mais comuns envolvidas nas ocorr√™ncias de disparos policiais fatais s√£o as armas de fogo, presentes em 2997 casos. Em seguida, as facas, presentes em 780 casos e armas de brinquedo est√£o presentes em 177 casos.
Vale mencionar que, apesar dessas serem as armas mais frequentes, tamb√©m ocorreram situa√ß√µes em que os suspeitos n√£o estavam armados (318 casos) e em que a presen√ßa de arma foi classificada como indeterminada (221 casos).

```python
# Qual √© a propor√ß√£o de suspeitos armados e desarmados nas ocorr√™ncias?

from sqlalchemy import create_engine
import pandas as pd

engine = create_engine('postgresql+psycopg2://postgres:123@localhost/postgres')
conn = engine.connect()

query = """
SELECT armed AS armado, COUNT(armed) AS total
FROM (
  SELECT CASE
   WHEN armed != 'unarmed' AND armed != 'undetermined'
  THEN 'armed'
  ELSE armed END AS armed
  FROM projeto.fatal_shootings
  ) AS subquery
GROUP BY armed
ORDER BY total DESC;
"""
df = pd.read_sql_query(query, con = conn)
df
```

A maioria dos casos envolve suspeitos armados com 4559 ocorr√™ncias, seguidos por 318 casos de suspeitos desarmados e 221 casos em que o status √© classificado como indeterminado.

Essas informa√ß√µes ressaltam a presen√ßa significativa de suspeitos armados nos casos analisados. No entanto, tamb√©m √© relevante notar que houve ocorr√™ncias em que os suspeitos estavam desarmados ou em que a classifica√ß√£o do status foi inconclusiva.

```python
# Quais cidades t√™m o maior n√∫mero de tiroteios fatais?

from sqlalchemy import create_engine
import pandas as pd

engine = create_engine('postgresql+psycopg2://postgres:123@localhost/postgres')
conn = engine.connect()

query = """
SELECT city AS cidade, COUNT(city) AS total
FROM projeto.fatal_shootings
GROUP BY city
ORDER BY total desc
LIMIT 5;
"""
df = pd.read_sql_query(query, con = conn)
df
```

```python
# Quais estados t√™m a maior incid√™ncia de tiroteios fatais?

from sqlalchemy import create_engine
import pandas as pd

engine = create_engine('postgresql+psycopg2://postgres:123@localhost/postgres')
conn = engine.connect()

query = """
SELECT state AS estado, COUNT(state) AS total
FROM projeto.fatal_shootings
GROUP BY state
ORDER BY total desc
LIMIT 5;
"""
df = pd.read_sql_query(query, con = conn)
df
```
